```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from math import pi,sqrt,cos,sin
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

#the gravitational constant
g = 1

ms = 1.2
me = 1.0
mm = 0.8

G = 1
```

```python
def gravity(r,t):
    sx = r[0]
    vsx = r[1]
    sy = r[2]
    vsy = r[3]

    ex = r[4]
    vex = r[5]
    ey = r[6]
    vey = r[7]

    mx = r[8]
    vmx = r[9]
    my = r[10]
    vmy = r[11]

    dsxdt = vsx
    dsydt = vsy
    dexdt = vex
    deydt = vey
    dmxdt = vmx
    dmydt = vmy

    dvsxdt = G * me * (ex-sx) / (((ex-sx)**2 + (ey-sy)**2) ** (3/2)) - G * mm * (sx-mx) / (((sx-mx)**2 + (sy-my)**2) ** (3/2))
    dvsydt = G * me * (ey-sy) / (((ex-sx)**2 + (ey-sy)**2) ** (3/2)) - G * mm * (sy-my) / (((sx-mx)**2 + (sy-my)**2) ** (3/2))

    dvexdt = - G * ms * (ex-sx) / (((ex-sx)**2 + (ey-sy)**2) ** (3/2)) + G * mm * (mx-ex) / (((mx-ex)**2 + (my-ey)**2) ** (3/2))
    dveydt = - G * ms * (ey-sy) / (((ex-sx)**2 + (ey-sy)**2) ** (3/2)) + G * mm * (my-ey) / (((mx-ex)**2 + (my-ey)**2) ** (3/2))

    dvmxdt = - G * me * (mx-ex) / (((mx-ex)**2 + (my-ey)**2) ** (3/2)) + G * ms * (sx-mx) / (((sx-mx)**2 + (sy-my)**2) ** (3/2))
    dvmydt = - G * me * (my-ey) / (((mx-ex)**2 + (my-ey)**2) ** (3/2)) + G * ms * (sy-my) / (((sx-mx)**2 + (sy-my)**2) ** (3/2))
    #x-position x-velocity y-position y-velocity
    ##################
    drdt = [dsxdt,dvsxdt,dsydt,dvsydt,dexdt,dvexdt,deydt,dveydt,dmxdt,dvmxdt,dmydt,dvmydt]

    return drdt
```

```python
#orbit 1
#initial condition
i0 = [0, 0.2, 0, -0.5, 1, 0.2, 1.2, -0.3, -0.3, 0.2, -1.2, -0.1]

k = 15
t = np.linspace(0,10,2**k+1)

ar = odeint(gravity,i0,t)
```

```python
#bounded orbit inside
fig, axl = plt.subplots(1,1,figsize=(8,8))

rsx = ar[:,0]
rsy = ar[:,2]

rex = ar[:,4]
rey = ar[:,6]

rmx = ar[:,8]
rmy = ar[:,10]

axl.plot(rsx, rsy, color='red', label='Position of Object 1')
axl.plot(rex, rey, color='blue', label='Position of Object 2')
axl.plot(rmx, rmy, color='green', label='Position of Object 3')

#the axis control function
#axl.set_ylim(-1.5,1.5)
#axl.set_xlim(-1.5,1.5)

axl.spines['left'].set_position('zero')

axl.spines['right'].set_color('none')
axl.spines['bottom'].set_position('zero')
axl.spines['top'].set_color('none')
axl.xaxis.set_minor_locator(MultipleLocator(0.05))
axl.yaxis.set_minor_locator(MultipleLocator(0.05))
axl.xaxis.set_tick_params(labelsize=10)
axl.yaxis.set_tick_params(labelsize=10)
axl.set_ylabel('y', fontsize=15)
axl.set_xlabel('x', fontsize=15)
axl.xaxis.set_label_coords(0.925, 0.475)
axl.yaxis.set_label_coords(0.475, 0.925)
axl.set_title('Three Body Problem Numerical Solution No.1',fontsize=14)
axl.legend(fontsize=12,loc='upper right')

plt.savefig('Three Body Problem Numerical Solution 1.png')
```
